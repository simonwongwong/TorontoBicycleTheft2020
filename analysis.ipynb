```python
import pandas as pd
import plotly.graph_objects as go
```

```python
crimes = pd.read_csv("Bicycle_Thefts.csv")
crimes
```

```python
crimes.columns
```

# Neighbourhoods

```python
nbhd = pd.DataFrame()
nbhd['neighbourhood'] = crimes['Neighbourhood'].apply(lambda n: n.split(" (")[0])
nbhd['thefts'] = crimes['ObjectId']
coord_df = pd.read_csv("neighbourhood_coord.csv", index_col=0)

joined_nbhd = nbhd.groupby("neighbourhood").count().sort_values('thefts').join(coord_df)
```

```python
graph = go.Bar(x=joined_nbhd.index, y=joined_nbhd.thefts)
n = go.Figure(graph)
n.update_layout(yaxis=go.layout.YAxis(title=go.layout.yaxis.Title(text="Number of Crimes")), xaxis={'range':[120,139.5]})
n.update_xaxes(tickangle=-50, tickmode='array')
n
```

```python
po.plot(n, filename='neighbourhoods.html')
```

```python
joined_nbhd['name'] = joined_nbhd.index
```

```python
import plotly.express as px
px.set_mapbox_access_token(open(".mapbox_token").read())
tmap = px.scatter_mapbox(joined_nbhd, lat="lat", lon="lon", color="thefts", size="thefts",text="name",
                  color_continuous_scale=px.colors.sequential.thermal, size_max=50, zoom=10)
tmap
```

```python
po.plot(tmap, filename='heatmap.html')
```

# Location Type

```python
lt = pd.DataFrame()
lt['location_type'] = crimes["Location_Type"]
lt['thefts'] = crimes["ObjectId"]
glt = lt.groupby('location_type').count()
```

```python
graph = go.Pie(labels=glt.index, values=glt.thefts, title="Location Type")
fig = go.Figure(graph)
fig.update_traces(textposition='inside')
fig.update_layout(uniformtext_minsize=12, uniformtext_mode='hide')
fig.show()
```

```python
import plotly.offline as po
po.plot(fig, filename="location_type.html")
```

# Time

```python
dtdf = pd.DataFrame()
dtdf['date'] = pd.to_datetime(crimes.Occurrence_Date)
dtdf['hour'] = [t[11:13] for t in crimes.Occurrence_Time]
dtdf['year'] = [t.year for t in dtdf.date]
dtdf
```

```python
hours = dtdf.groupby("hour").count()[['date']]
hours.columns = ['thefts']
graph = go.Bar(x=hours.index, y=hours.thefts)
ht = go.Figure(graph)
ht.update_layout(yaxis=go.layout.YAxis(title=go.layout.yaxis.Title(text="Number of Crimes")),
                  xaxis=go.layout.XAxis(title=go.layout.xaxis.Title(text="Hour of Day")))
ht.update_xaxes(tickmode='linear')
ht
```

```python
po.plot(ht, filename='hourly_thefts.html')
```

```python
yearly = dtdf.groupby("year").count()[['date']]
yearly.columns = ['thefts']
graph = go.Bar(x=yearly.index, y=yearly.thefts)
yt = go.Figure(graph)
yt.update_layout(yaxis=go.layout.YAxis(title=go.layout.yaxis.Title(text="Number of Thefts")),
                  xaxis=go.layout.XAxis(title=go.layout.xaxis.Title(text="Year")))
yt
```

```python
po.plot(yt, filename='yearly_thefts.html')
```

```python
monthly = dtdf.set_index('date').resample('M').count()[['hour']]
monthly.columns = ['thefts']

graph = go.Bar(x=monthly.index, y=monthly.thefts)
m = go.Figure(graph)
m.update_layout(yaxis=go.layout.YAxis(title=go.layout.yaxis.Title(text="Number of Thefts")),
                  xaxis=go.layout.XAxis(title=go.layout.xaxis.Title(text="Month")))
m
```

```python
po.plot(m, filename='monthly_thefts.html')
```
